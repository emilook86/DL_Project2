```python
import os
import random
import shutil

################################## SET THIS! ######################################
PATH = "data_copy"  # path to dataset
COPY = True  # set to True if you want to create a copy of data before splitting
COPY_PATH = "../data_copy"  # path to copy of data
###################################################################################

# create copy of data in case when something goes wrong
if COPY and not os.path.exists(COPY_PATH):
    shutil.copytree(PATH, COPY_PATH)

labels = os.listdir(PATH)
if (
    not os.path.exists(f"{PATH}/train")
    or not os.path.exists(f"{PATH}/val")
    or not os.path.exists(f"{PATH}/test")
):
    os.makedirs(f"{PATH}/train")
    os.makedirs(f"{PATH}/val")
    os.makedirs(f"{PATH}/test")
for label in labels:
    if not os.path.exists(f"{PATH}/train/{label}"):
        os.makedirs(f"{PATH}/train/{label}")
    if not os.path.exists(f"{PATH}/val/{label}"):
        os.makedirs(f"{PATH}/val/{label}")
    if not os.path.exists(f"{PATH}/test/{label}"):
        os.makedirs(f"{PATH}/test/{label}")

# split user ids into train, val, test - we ensure that each user is in only one set
user_ids = set()
for label in labels:
    files = os.listdir(f"{PATH}/{label}")
    for file in files:
        user_id = file.split("_")[0]
        user_ids.add(user_id)
user_ids = list(user_ids)
random.Random(42).shuffle(user_ids)
train_ids = user_ids[: int(0.7 * len(user_ids))]
val_ids = user_ids[int(0.7 * len(user_ids)) : int(0.8 * len(user_ids))]
test_ids = user_ids[int(0.8 * len(user_ids)) :]

# move files to train, val, test folders
for label in labels:
    files = os.listdir(f"{PATH}/{label}")
    train = [file for file in files if file.split("_")[0] in train_ids]
    val = [file for file in files if file.split("_")[0] in val_ids]
    test = [file for file in files if file.split("_")[0] in test_ids]
    for file in train:
        os.rename(f"{PATH}/{label}/{file}", f"{PATH}/train/{label}/{file}")
    for file in val:
        os.rename(f"{PATH}/{label}/{file}", f"{PATH}/val/{label}/{file}")
    for file in test:
        os.rename(f"{PATH}/{label}/{file}", f"{PATH}/test/{label}/{file}")

# delete empty folders PATH/label
for label in labels:
    os.rmdir(f"{PATH}/{label}")
```
